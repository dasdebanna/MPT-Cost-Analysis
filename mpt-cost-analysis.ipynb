```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for Training and Fine-tuning Costs
models = ['MPT']
training_costs = [0.0005]
finetuning_costs = [0.001]
```

```python
# Plot Training and Fine-tuning Costs
fig, ax = plt.subplots()
bar_width = 0.35
index = range(len(models))

bar1 = plt.bar(index, training_costs, bar_width, label='Training Cost')
bar2 = plt.bar([i + bar_width for i in index], finetuning_costs, bar_width, label='Fine-tuning Cost')

plt.xlabel('Models')
plt.ylabel('Cost per 1,000 tokens ($)')
plt.title('Training and Fine-tuning Costs for MPT')
plt.xticks([i + bar_width / 2 for i in index], models)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Data for Usage Costs (Input and Output)
input_costs = [0.0005]
output_costs = [0.001]
```

```python
# Plot Usage Costs
fig, ax = plt.subplots()
index = np.arange(len(models))

bar3 = plt.bar(index, input_costs, bar_width, label='Input Cost')
bar4 = plt.bar(index + bar_width, output_costs, bar_width, label='Output Cost')

plt.xlabel('Models')
plt.ylabel('Cost per 1,000 tokens ($)')
plt.title('Usage Costs for MPT')
plt.xticks(index + bar_width / 2, models)
plt.legend()

plt.tight_layout()
plt.show()
```
